```python
# find that yaml file
config_path = None
for f in glob('./*.yaml'):
    config_path = f
    break
```

```python
config_path
```

```python
import yaml

subfolder_name = None
with open(config_path) as cfg:
	d = yaml.load(cfg, Loader=yaml.FullLoader)
	subfolder_name = d['exp_config']['experiment_name'].split(',')[0]
```

```python
exp_path = os.path.join('/content/drive/MyDrive/phlk_exp/', subfolder_name)
try:
    os.mkdir(exp_path)
except:
    pass
```

```python
runs_path = os.path.join(exp_path, 'runs')
try:
    os.mkdir(runs_path)
except:
    pass
```

```python
path_to_teacher = None
for f in glob(os.path.join(exp_path, 'teacher*.pt')):
    path_to_teacher = f
    break

# if unsuccessful to load from disk, then try getting the downloaded one
if path_to_teacher is None:
    for f in glob('./teacher*.pt'):
        path_to_teacher = f
        break
```

```python
path_to_student = None
'''
for f in glob(os.path.join(exp_path, 'student*.pt')):
    path_to_student = f
    break
'''
```

```python
from ADR_crosslingual.trainers.main import main

teacher_save_path = None
student_save_path = None

# если на диске есть место, лучше раскомментируйте вот эти две строчки
#teacher_save_path = os.path.join(exp_path, 'teacher_chk.pt')
#student_save_path = os.path.join(exp_path, 'student_chk.pt')

writer = main(
    config_path, runs_path,
    teacher_load_path=path_to_teacher, teacher_save_path=teacher_save_path,
    student_load_path=path_to_student, student_save_path=student_save_path,
    do_train_teacher=False, do_train_student=True,
)
```
